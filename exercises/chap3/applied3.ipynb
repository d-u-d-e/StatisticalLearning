Exercise 8: College dataset

```python
college = pd.read_csv("datasets/College.csv")
```

```python
college2 = pd.read_csv("datasets/College.csv", index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3 = college3.set_index('College')
```

```python
college2
```

```python
college3
```

```python
college = college3
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']])
```

```python
college.boxplot('Outstate', by='Private')
```

```python
college['Elite'] = pd.cut(college['Top10perc'], [0, 0.5, 1], labels=['No', 'Yes'])
```

```python
college['Elite'].value_counts()
```

```python
college.boxplot('Outstate', by='Elite')
```

```python
fig, ax = plt.subplots(figsize=(8,8), nrows=2, ncols=2)
college.plot.hist(column='Enroll', bins=6, ax=ax[0, 0])
college.plot.hist(column='Apps', bins=6, ax=ax[0, 1])
college.plot.hist(column='Accept', bins=6, ax=ax[1, 0])
college.plot.hist(column='Grad.Rate', bins=6, ax=ax[1, 1])
```
